```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import glob

# Load images
lmd=0.5000
image_files = sorted(glob.glob(f"/home/preuer/OneDrive/Studium/Master/Semester/Sem_4/Master Thesis/fairness_triangle/img/E1_Label_Noise_for_Single_Points/Flipping_Probability_Effect_on_Corruption_lmd_{lmd}*.png"))  # Adjust path

# Create figure
fig, ax = plt.subplots()
ax.axis("off")  # Hide axis
img = plt.imshow(plt.imread(image_files[0]))  # Load first image

# Update function
def update(frame):
    img.set_array(plt.imread(image_files[frame]))
    return img,

# Create animation
ani = animation.FuncAnimation(fig, update, frames=len(image_files), interval=500, blit=True)

# Save as GIF (or use plt.show() to view it live)
ani.save(f"img/gifs/Flipping_prob_Effect_Corruption_lmd_{lmd}.gif", writer=animation.PillowWriter(fps=3))

plt.show()
```
