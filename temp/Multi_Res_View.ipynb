```python
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(["science"])
plt.rcParams.update({
    'font.size': 14,
    'axes.labelsize': 14,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14
})

# Create the figure and axis
fig, ax = plt.subplots(figsize=(2, 6))

# Draw the vertical line
ax.plot([0.5, 0.5], [0, 1], color="black", linewidth=2)

# Define positions and labels
positions = [1.0, 0.75, 0.0]  
labels = ["Coars Group", "Nationalities", "Individuals"]

# Very short tick length
tick_half_length = 0.001  # total width = 0.002, much smaller

# Draw centered ticks (tiny) and labels close to the line
for pos, label in zip(positions, labels):
    ax.plot([0.5 - tick_half_length, 0.5 + tick_half_length], [pos, pos],
            color="black", linewidth=1.2)  
    ax.text(0.505, pos, label, ha="left", va="center", fontsize=13)

# Bold the top level
ax.text(0.505, positions[0], labels[0], ha="left", va="center",
        fontsize=13, fontweight="bold")

# Title of the illustration
ax.set_title("Multi-Resolution View", fontsize=15, pad=20)

# Remove axes
ax.axis("off")

# Save as vector for thesis


plt.show()
```
