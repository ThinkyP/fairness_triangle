```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scienceplots
plt.style.use(['science'])   # Activate science plot style

# Define fairness values
tau = np.linspace(0, 0.25, 300)

# Define 2 risk functions
a_1 = 1.6
a_2 = 1.8
risk1 = a_1 * tau**2                     # blue
risk2 = a_2 * tau**2 + 0.025             # black

# Fixed y = 0.05 for horizontal line
y_fixed = 0.05

# Solve for x positions where each curve reaches y = 0.05
x1_cross = np.sqrt(y_fixed / a_1)
x2_cross = np.sqrt((y_fixed - 0.025) / a_2)

# Plot
plt.figure(figsize=(5.5, 3.5))
plt.plot(tau, risk1, color='blue', label='pareto frontier')
plt.plot(tau, risk2, color='black', label='corr pareto frontier')

# Vertical line at x = 0.10
x_val = 0.10
y1_val = a_1 * x_val**2
y2_val = a_2 * x_val**2 + 0.025
plt.vlines(x=x_val, ymin=y1_val, ymax=y2_val, colors='red', linestyles='dotted', label=r'$\tau = 0.10$')

# Horizontal line between both curves at y = 0.05
plt.hlines(y=y_fixed, xmin=x2_cross, xmax=x1_cross, colors='red', linestyles='dotted', label=r'$F(\tau)=0.05$')

# Axes labels
plt.xlabel(r"$\tau$",  fontsize=14)
plt.ylabel(r"$F(\tau)$",  fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Limits
plt.xlim(0, 0.25)
plt.ylim(0, 0.21)

# Legend
plt.legend(loc='upper right', fontsize=12)

from matplotlib.ticker import FormatStrFormatter
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))


# Finalize
plt.tight_layout()
plt.show()
```
